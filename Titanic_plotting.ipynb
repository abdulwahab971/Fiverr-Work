# More Complex histograms

We can use the hue attribute of histograms and distribution plots to have more complex histogram. On the example below, we use hue= 'sex' to view within the **fare** histrogram, how many of those fares were purchased by man and how many of those by woman

```python
sns.histplot(data=titanic_df, x="fare", hue = "sex")
```

In the previous two histogram matplotlib automatically chose the number of bins for the histogram, we can manually set the number of bins to create a more readable histogram by using the bins parameter. In the examples below, we are using **bins=15** to indicate that the histogram should have 15 bars.

```python
sns.histplot(data=titanic_df, x="fare", bins = 15)
```

```python
sns.histplot(data=titanic_df, x="fare", hue = "sex", bins = 15)
```

Notice that the labels on the y-axis changed when we used the hue parameter. This is because matplotlib is now creating two bars for each bins (one for male and one for female). 
As you can see, both bars start at the bottom of the y-axis with the orange bars being shown behind the blue bars

This defeats the purpose of the plot, since it makes it hard to visualize the proportion of man and woman. We can use the attribute multiple to indicate how to handle multiple bars. Is we set the attribute to "stack", the bars will be shown one on top of the other as shown below. 

Furthermore, we can also show the bars side by side by setting the multiple attribute to "dodge"

```python
sns.displot(data=titanic_df, x="fare", hue="sex", multiple="stack", bins=15)
```

# Task g) Dodge

Create the same histogram as the example above but have the bars for man and woman shown side-by-side

```python
# Write code for task g on this cell
import seaborn as sns

# create separate dataframes for male and female passengers
male_df = titanic_df[titanic_df['sex'] == 'male']
female_df = titanic_df[titanic_df['sex'] == 'female']

# create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# create a histogram of the "fare" column for male passengers
sns.histplot(male_df['fare'], bins=20, ax=axs[0], color='blue')

# create a histogram of the "fare" column for female passengers
sns.histplot(female_df['fare'], bins=20, ax=axs[1], color='red')


# display the plot
plt.show()
```
